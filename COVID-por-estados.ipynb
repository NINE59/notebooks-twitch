```julia
using Colors, Krylov

states = unique(data.state)
st_idx = findall([data[data.state .== st,:cases][end] for st in states] .>= 100)
states = states[st_idx]
clrs = distinguishable_colors(length(states) + 1, [RGB(0, 0, 0), RGB(1.0, 0.0, 0.0)])[2:end]

plot(size=(800,400))
for (i,st) in enumerate(states)
    df = data[data.state .== st,:]
    #print(df)
    I = setdiff(unique([findfirst(df.date .== dt) for dt in dates]), [nothing])
    #print(I)
    df = df[I,:]

    idx = findfirst(df.cases .≥ 100)
    y = df.cases[idx:end]
    n = length(y)
    x = 1:n

    X = [ones(n) x]
    yhat = log.(y)

    β = Krylov.cgls(X, yhat)[1]
    
    dates_idx = [findfirst(dates .== d) for d in df.date[idx:end]]

    plot!(dates_idx, y, m=:dot, ms=5, lab="", c=clrs[i])
    next = 1
    xnext = (1:n+next)
    append!(dates_idx, dates_idx[end] .+ (1:next))
    plot!(dates_idx, exp.(β[1] .+ β[2] * xnext), c=clrs[i], lw=2, yaxis=:log, m=:dot, ms=2,
          lab="$st rate: $(round(100β[2], digits=2))%")
end
plot!()
```

```julia
using Logging, NLPModels, NLPModelsIpopt, JSOSolvers

st = "São Paulo"

df = data[data.state .== st,:]
#print(df)
I = setdiff(unique([findfirst(df.date .== dt) for dt in dates]), [nothing])
#print(I)
df = df[I,:]

idx = findfirst(df.cases .≥ 100)
y = df.cases[idx:end]
n = length(y)
x = 1:n

h(β, xi) = β[3] / (1 + exp(-β[1] - β[2] * xi))
lvar = [-Inf; -Inf; y[end] * 2]
uvar = [ 0.0;  Inf;        Inf]

res(β) = [y[i] - h(β, x[i]) for i = 1:n]
nls = ADNLSModel(res, [-1.0; 1.0; y[end] * 3], n, lvar=lvar, uvar=uvar)
output = with_logger(NullLogger()) do
    tron(nls)
end
β = output.solution
print(output)
```

```julia
using Logging, NLPModels, NLPModelsIpopt, JSOSolvers

states = unique(data.state)
st_idx = findall([data[data.state .== st,:cases][end] for st in states] .>= 100)
states = states[st_idx]
clrs = distinguishable_colors(length(states) + 1, [RGB(0, 0, 0), RGB(1.0, 0.0, 0.0)])[2:end]

plot(size=(800,300))

for (i,st) in enumerate(states)
    df = data[data.state .== st,:]
    #print(df)
    I = setdiff(unique([findfirst(df.date .== dt) for dt in dates]), [nothing])
    #print(I)
    df = df[I,:]

    idx = findfirst(df.cases .≥ 100)
    y = df.cases[idx:end]
    n = length(y)
    x = 1:n

    h(β, xi) = β[3] / (1 + exp(-β[1] - β[2] * xi))
    lvar = [-Inf; -Inf; y[end] * 2]
    uvar = [ 0.0;  Inf;        Inf]

    res(β) = [y[i] - h(β, x[i]) for i = 1:n]
    nls = ADNLSModel(res, [-1.0; 1.0; y[end] * 3], n, lvar=lvar, uvar=uvar)
    output = with_logger(NullLogger()) do
        #tron(nls)
        ipopt(nls)
    end
    β = output.solution
    
    dates_idx = [findfirst(dates .== d) for d in df.date[idx:end]]

    next = 7
    xnext = (1:n+next)
    
    plot!(dates_idx, y, m=:dot, ms=5, lab="", c=clrs[i])
    
    next = 1
    xnext = (1:n+next)
    append!(dates_idx, dates_idx[end] .+ (1:next))
    hx = [h(β, xi) for xi in xnext]
    plot!(dates_idx, hx, c=clrs[i], lw=2, yaxis=:log, m=:dot, ms=2,
          lab="$st max: $(β[3])")
end
plot!()
```
